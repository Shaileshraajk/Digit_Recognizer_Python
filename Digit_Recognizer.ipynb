```python
import pandas as pd
df = pd.read_csv(Path+'/digit_svm.csv')
df
```

# Checking for null values

```python
df.isnull().sum()
```

# Separating features and target variable



```python
x = df.drop(["label"],axis="columns")
y = df["label"]
```

```python
x
```

```python
y
```

# Splitting dataset into training data and test data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=  0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Training the Data using Random Forest Classification Model


```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=200)
model.fit(x_train,y_train)
```

# Testing the Data

```python
y_pred=model.predict(x_test)
y_predl = list(model.predict(x_test))
test_df = {'Predicted':y_predl,'Actual':y_test}
pd.DataFrame(test_df).head(10)
```

# Evaluation of the Model - Finding Accuracy and Error value

```python
from sklearn.metrics import accuracy_score, confusion_matrix, mean_squared_error

mae=abs(y_pred-y_test).mean()
print("Mean Absolute Error: ",mae)

mse = mean_squared_error(y_test,y_pred)
print('Mean Squared Error: ',mse)

acc = accuracy_score(y_test,y_pred)
acc = "{:.2%}".format(acc)
print('Accuracy: ',acc)
```
